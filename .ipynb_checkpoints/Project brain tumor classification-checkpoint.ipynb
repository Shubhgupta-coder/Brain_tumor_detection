```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
```

# Prepare collect data


```python
pip install opencv-python
```

```python
import os
path=os.listdir('Training/')
classes= {'no_tumor':0 , 'pituitary_tumor':1,'meningioma_tumor':2,'glioma_tumor':3}
```

```python
import cv2
X=[]
Y=[]
# in first iterartion in Training data we have no_tumor and in second iteration we have pituitary_tumor
for cls in classes:
    pth='Training/'+cls
    #in order to read all the images we have done nestrd loop
    for j in os.listdir(pth):         
        img= cv2.imread(pth+'/'+j,0)  #in order to read all the image in 2 dimension
        img=cv2.resize(img,(200,200)) #in order to have all images of same size
        X.append(img)
        Y.append(classes[cls])
    
```

```python
np.unique(Y) #it show us two classes 0 and 1 
```

```python
X=np.array(X)
Y=np.array(Y)
```

```python
pd.Series(Y).value_counts()  #to check the no. of no tumor and pituitary tumor images in training dataset
```

```python
X.shape
```

# visualise data


```python
plt.imshow(X[77],cmap='gray')    #this is to show each image as dataset in 2D array
```

```python
X_updated=X.reshape(len(X),-1) #because sklearn work on only 2 dimension
X_updated.shape
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X_updated,Y,test_size=0.20,random_state=10)#we take 20%data for testing and remaining data for training 
```

```python
xtrain.shape,xtest.shape
```

# Feature scaling

```python
print(xtrain.max(),xtrain.min())  #to normailse the data within a particular range
print(xtest.max(),xtest.min())
xtrain=xtrain/255
xtest=xtest/255
print(xtrain.max(),xtrain.min())
print(xtest.max(),xtest.min())
```

```python
from sklearn.decomposition import PCA
pca=PCA(.98)
pca_train=xtrain
pca_test=xtest
```

# Train Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
import warnings
warnings.filterwarnings('ignore')
lg=LogisticRegression(C=0.1)
lg.fit(pca_train, ytrain)
```

```python
sv=SVC()
sv.fit(pca_train,ytrain)
```

```python
print("training score:",lg.score(pca_train, ytrain))
print("testing score:",lg.score(pca_test, ytest))
```

```python
print("training score:",sv.score(pca_train, ytrain))
print("testing score:",sv.score(pca_test, ytest))
```

```python
#sv score is better than lg score
```

# prediction

```python
pred=sv.predict(pca_test)
np.where(ytest!=pred) #these are those indexes where data are not accurate
```
